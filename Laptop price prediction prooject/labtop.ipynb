```python
new = df['ScreenResolution'].str.split('x',n=1,expand=True)
```

```python
df['X_res'] = new[0]
df['Y_res'] = new[1]
```

```python
## GET X_axis resolutions only not get the any string values
df['X_res'] = df['X_res'].str.replace(',','').str.findall(r'(\d+\.?\d+)').apply(lambda x: x[0])
```

```python
# Change datatype of X_res,y_res into int
df['X_res'] = df['X_res'].astype('int64')
df['Y_res'] = df['Y_res'].astype('int64')
```

```python
# check the relation with Price
df.corr()['Price']
```

```python
# make a new column PPI by using these three columns X_res,y_res,Inches etc
df['PPI'] = (((df['X_res']**2)+(df['Y_res']**2))**0.5 /df['Inches'])
```

```python
# we do feature engineering and make one column by using these three column so there is not need 
df.drop(columns={'ScreenResolution','Inches','X_res','Y_res'},inplace=True)
```

```python
df['Cpu'].value_counts()
```

```python
df['Cpu'].apply(lambda x:x.split()[0:3])
```

```python
# remove , between  [Intel, Core, i5] and make a string by using " ",join
df['Cpu_Name'] = df['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
def processor_name(text):
    if text == 'Intel Core i3' or text == 'Intel Core i5' or text == 'Intel Core i7':
        return text
    else:
        if text.split()[0] == 'Intel':
            return 'other Intel Processor'
        else:
            return 'AMD Processor'
```

```python
df['Cpu_brand'] = df['Cpu_Name'].apply(processor_name)
```

```python
df.drop(columns = {'Cpu_Name','Cpu'},inplace=True)
```

```python
df['Cpu_brand'].value_counts().plot(kind='bar',color='skyblue')
```

**Memory Analysis**

```python
df['Memory'].value_counts()
```

```python
df['Memory'] = df['Memory'].astype(str).replace('\.0','',regex= True)
df['Memory'] = df['Memory'].str.replace('GB','')
df['Memory'] = df['Memory'].str.replace('TB','000')
new = df['Memory'].str.split('+',n=1,expand =True)

df['first'] = new[0]
df['first']=df['first'].str.strip()

df['second']= new[1]

df['Layer1HDD'] = df['first'].apply(lambda x:1 if x in 'HDD' else 0 )
df['Layer1SSD'] = df['first'].apply(lambda x:1 if x in 'SSD' else 0)
df['Layer1Hybrid']= df['first'].apply(lambda x:1 if x in 'Hybrid' else 0)
df['Layer1Flash_Storage'] = df['first'].apply(lambda x:1 if x in 'Flash Storage' else 0)

df['first'] = df['first'].str.replace(r'\D', '')
df['second'].fillna('0',inplace=True)

df['Layer2HDD'] = df['second'].apply(lambda x:1 if x in 'HDD' else 0)
df['Layer2SDD'] = df['second'].apply(lambda x:1 if x in 'SDD' else 0)
df['Layer2Hybrid'] = df['second'].apply(lambda x:1 if x in 'Hybrid' else 0)
df['Layer2Flash_Storage'] = df['second'].apply(lambda x:1 if x in 'Flash Storage' else 0)

df['second'] = df['second'].str.replace(r'\D', '')

df['first']=df['first'].astype('int64')
df['second']=df['second'].astype('int64')
```
